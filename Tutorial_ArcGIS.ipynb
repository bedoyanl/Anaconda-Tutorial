## First, import libraries for arcgis

```python
from arcgis.gis import GIS
from arcgis import features
from IPython.display import display
from arcgis.features import FeatureLayerCollection
from arcgis.features import FeatureLayer
```

## Create GIS object by creating a variable called gis and set equal to GIS() which is from the first line of the library import


```python
gis = GIS()
```

## This line of code creates map of los angeles.  Create a new variable called map_test and set it equal to gis which you defined above call the arcgis method "map" on the gis variable to create a new map of LA

```python
map_test = gis.map('Los Angeles')
```

## Time to add layers! 
### So, we have gotten our base map, but now we want to add layers. Find free layers on arcgis.com.  Or you can do a generic search on google. Or querey results seen further on.  In this example, we use a public feature service layers for USFW below. To look at the feature layer on arcgis.com go to this website below
### https://www.arcgis.com/home/item.html?id=8424d1c6ed314247a53efcc64e2edb08
### Copy the item id from the url it is the number following id=8424d1c6ed314247a53efcc64e2edb08
### This format will be the same for every url you have of a feature layer. Put the item id found in the url in the line below to retrieve the feature layer. Here I am creating a new variable called new_item that will retirve the feature layer by the item id. We are calling the gis variable like we did when we made the map

```python
wildlife_land = gis.content.get("8424d1c6ed314247a53efcc64e2edb08")
wildlife_land
```

## You can also search for feature layers here by adjusting the querey and item type
### Here, we are searching for layers that have Los Angeles in the title

```python
search_result_la = gis.content.search(query="title:Los Angeles*", outside_org=True, item_type = "Feature Layer")                  
search_result_la
```
